```python
import configparser
import os


def get_api_file_path():
    config = configparser.ConfigParser()
    script_path = os.path.dirname(os.path.abspath(__file__))
    config_file_path = os.path.join(script_path, 'config.ini')
    config.read(config_file_path)
    api_link = config['API']['api_link']
    return api_link



def get_MongoDB_info():
    config = configparser.ConfigParser()
    script_path = os.path.dirname(os.path.abspath(__file__))
    config_file_path = os.path.join(script_path, 'config.ini')
    config.read(config_file_path)
    mongodb_uri = config['MONGODB']['mongodb_uri']
    database_name= config['MONGODB']['database_name']
    collection_name= config['MONGODB']['collection_name']
    return mongodb_uri,database_name,collection_name
```

```python
import requests
import logging
from logger import LogHandler
from exception import CustomException

loghandler=LogHandler()
loghandler.setup_logging()



def get_weather(api_link):
    try:
        response = requests.get(api_link)
        response.raise_for_status()
        weather_data = response.json()
        logging.info(f"Got weather data from the API")
        # print(weather_data)
        return weather_data
    except requests.exceptions.RequestException as e:
        logging.error(f"Error occurred: {e}")
        return None
    
def is_valid_document(document):
    return isinstance(document, dict)    
    

def save_to_mongodb(mongo_client, database_name, collection_name, data):
    db = mongo_client[database_name]
    collection = db[collection_name]
    for d in data:
        if is_valid_document(d):
            collection.insert_one(d)
        else:
            print("Invalid document:", d)



```

```python
import configparser
import logging
from logger import LogHandler
from weather_api import WeatherDataProcessor
import pymongo
from utils import ConfigManager

# To handle logging
loghandler=LogHandler()
loghandler.setup_logging()

#To read the datas from config.ini
config_manager = ConfigManager()

# Get API file and the Mongodb details from config.ini using ConfigManager
api_link = config_manager.get_api_file_path()
mongodb_uri, database_name, collection_name = config_manager.get_MongoDB_info()

# To get the datas from API and save to MongoDB
weather_data_api=WeatherDataProcessor(api_link, mongo_client, database_name, collection_name)


def get_weather_save_db():
  
    # MongoDB configuration
    mongo_client = pymongo.MongoClient(mongodb_uri)

    # Get weather data from the API
    weather_data = weather_data_api.get_weather(api_link)

    if weather_data:
        # Save data to MongoDB
        weather_data_api.save_to_mongodb(mongo_client, database_name, collection_name, weather_data)

        # Log a success message
        logging.info("Weather data saved to MongoDB.")
    else:
        # Log a failure message

        logging.warning("Weather data not available.")


if __name__ == "__main__":
    get_weather_save_db()
```
